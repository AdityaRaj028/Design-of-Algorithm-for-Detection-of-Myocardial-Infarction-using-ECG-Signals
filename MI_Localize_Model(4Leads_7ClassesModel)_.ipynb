```python
# Loading the train and test data 
X_train = np.load('/content/Pdrive/MyDrive/BTP/XTrain.npy')
X_test = np.load('/content/Pdrive/MyDrive/BTP/XTest.npy')
y_train = np.load('/content/Pdrive/MyDrive/BTP/yTrain.npy')
y_test = np.load('/content/Pdrive/MyDrive/BTP/yTest.npy')
```

```python
# Import all the necessary things for model-training.

import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Activation, InputLayer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
    model = tf.keras.Sequential()
    model.add(InputLayer(input_shape = (48,)))
    model.add(Dense(units=70,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(units=70,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(units=70,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(units=7, activation="softmax"))
    
```

```python
model.summary()
```

```python
model.compile(optimizer = Adam(learning_rate = 0.0001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.fit(x = X_train, y = y_train, batch_size = 256, epochs = 200)
```

```python
score = model.evaluate(X_test, y_test, batch_size = 256)
```

```python
# THE FINAL ACCURACY RESULT OF THE MODEL:

print("Accuracy: ", round(score[1]*100, 2))
```
